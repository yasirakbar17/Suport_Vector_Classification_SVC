```python
df.Beverage_category = df.Beverage_category.apply(lambda x : int(1) if x == 'Classic Espresso Drinks' else int(0) )
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('Beverage_category', axis = 1).values
y = df.Beverage_category.values

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = df.Beverage_category, random_state = 42)
```

```python
scale = StandardScaler()
scale.fit(X_train)
X_train_scale = scale.transform(X_train)
X_test_scale = scale.transform(X_test)
```

```python
svm = SVC(C=0.5, probability=True, verbose=1, random_state = 12 )
model = svm.fit(X_train_scale, y_train)
```

```python
print(f'Akurasi Data Train : {model.score(X_train_scale, y_train):.3f}')
print(f'Akurasi Data Test : {model.score(X_test_scale, y_test):.3f}')
```

```python
pred_proba_train = pd.DataFrame(model.predict_proba(X_train_scale), columns = ['0', '1'])
pred_proba_train.head()
```

```python
pred_proba_test = pd.DataFrame(model.predict_proba(X_test_scale), columns = ['0', '1'])
pred_proba_test.head()
```
